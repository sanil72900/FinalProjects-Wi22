**Relevant Prior Work: What has already been done on this topic?**
        	
There have been many studies that acknowledge the detrimental effects of alcohol on individuals. However, there have been a handful of studies that specifically examine the negative repercussions of alcohol on high school students and their well-being and education. In a study originally published in the Journal of Community Psychology, the correlation between “alcohol outlet density, perceived alcohol, tobacco, and marijuana availability (ATOD), perception of substance use as a problem at the school, and self-reported ATOD use” was observed. The study surveyed 25,308 students in the participating 58 high schools in Maryland and utilized this data in conjunction with alcohol outlet data from the Liquor License Boards (1). Within this study, the researchers found that there was a positive correlation between the alcohol outlet density and perceived availability of substances amongst girls specifically (1). Although this study surveyed a large sample of students, it must be acknowledged that the survey may not provide the most accurate results, considering students self-reported their behaviors and intentions. Even more, this study did not necessarily delve into the effects of alcohol use on academic performance and high school graduation rates.

Another study observed the correlation between the presence of alcohol outlets in a region to the alcohol use of high school students (2). Specifically, the researchers in this study observed 242 high school students within the metropolitan area of Minneapolis-St. Paul, Minnesota. Students completed a survey to assess their alcohol use and the distance between the alcohol outlet and the student’s home and school was calculated using Geographic Information System (GIS) data. Researchers within this specific study established that there was no correlation between the exposure to alcohol outlets in a region to the alcohol use of high school students within this specific area (2). Similar to the previous study, this study utilized surveys amongst the students which could provide inaccurate information. Additionally, the study did not focus on the graduation rates of the students or the possible impact on academic performance of the students.

However, there was a specific study (3) that was done to observe the relationship between academic achievement and alcohol use that especially caught our eye. Within this study, researchers first surveyed 90,118 students amongst 175 schools between 1994-1995. Next, researchers chose to interview 20,745 students and their parents from the initial pool of students in their homes, and followed up one year later with a similar interview. In 2001–2002, students from the previous interviews that were now 18-26 years of age were re-interviewed to further observe the repercussions of adolescent drinking and substance use on their lives later. All of the respondents of the study were asked to allow a request for official transcripts, and the study assessed the GPA of students from these transcripts. Through this study, researchers found that “increases in alcohol consumption result in small yet statistically significant reductions in GPA for male students and in statistically non-significant changes for females. For females, however, higher levels of drinking result in self-reported academic difficulty” (3). This study is similar to our research question, and assessed the correlation between alcohol use and academic achievement. Although the study does not exactly observe the same relationship we wish to investigate, the results help us assess that our research question is relevant to delve deeper into.

```python
sns.set(font_scale = 1.5)
g =sns.lmplot(y= 'Graduation_Percentage',
          x= 'Density_of_Alcohol_Outlets',
          data= density_outlets_avg_grad,
          fit_reg= True,
          height= 6,
          aspect= 2,
          line_kws= {'color': 'blue'},
          scatter_kws= {'color': 'black'})
var = g.set(title = 'Density of Alcohol Outlets vs. Graduation Percentage')
```

From the regression plot above there is no remarkable correlation between the `Density_of_Alcohol_Outlets` and the `Graduation_Percentage` as the regression line has a slope of approximately zero. However, we can further our understanding by also doing an OLS Regression.

We will conduct an OLS Regression to check if our observations from the regression line plot above are valid.

```python
dependant, predictors = patsy.dmatrices('Graduation_Percentage ~ Density_of_Alcohol_Outlets', density_outlets_avg_grad)
mod = sm.OLS(dependant, predictors)
res = mod.fit()
print(res.summary())
```

**OLS Analysis:** The p-value from the OLS Regression Results above is 0.891 which is greater than 0.05 (we are taking the 95% confidence interval with alpha = 0.05). Therefore the results are not statistically significant. We accept our null hypothesis that there is no correlation between the density of alcohol outlets and the high school graduation percentage, and reject our alternate hypothesis. Additionally, the R-squared value is 0.000 which is far from 1.0, indicating that our results are not statistically significant.

To further establish our findings, we will also conduct a Pearson's correlation test.

```python
print('model beta: ' , res.params[1])
print('Pearson\'s correlation: ', stats.pearsonr(density_outlets_avg_grad['Density_of_Alcohol_Outlets'], density_outlets_avg_grad['Graduation_Percentage'])[0])
```

The Pearson's correlation yields 0.01953743066075868, which indicates that our results are not statistically significant.
