#### introduction

However, if employees don’t just leave because they have to. They may be pushed, whether intentionally or because there’s something in the business that isn’t quite right. According to the 2021 Bureau of Labor Statistics report, 4 million people quit their job in April and another 4 million people quit in July (1). Thus, in our study, we are attempting to examine how employees’ personal factors or contextual factors affect their tendency of attrition and figure out which one is the most influential factor: personal or contextual. Based on the boxplot, we notice that there are outliers for attrition-employee ages. So let's find them.

```python
def outliers(arr):
    # finding the 1st quartile
    q1 = np.quantile(arr, 0.25)
    
    # finding the 3rd quartile
    q3 = np.quantile(arr, 0.75)
    med = np.median(arr)
 
    # finding the iqr region
    iqr = q3-q1
 
    # finding upper and lower whiskers
    upper_bound = q3+(1.5*iqr)
    lower_bound = q1-(1.5*iqr)
    outliers = arr[(arr <= lower_bound) | (arr >= upper_bound)]
    
    return outliers
#find the outliter for attrition-employee agres
age_attrition_outliers = df[df['Attrition'] == 'Yes'].get('Age')
outliers(age_attrition_outliers)
```

We found that there are 8 outliers in the attrition-employee age. Detailed outliers are shown above.

### Personal factor:

#### Age (personal factor) with attrition
- Hypothesis Testion
  
  $H_{o}$: There is no relationship between age and employee attrition
  
  $H_{A}$: There is relationship between age and employee attrition

```python
outcome_1, predictors_1 = patsy.dmatrices('Age ~Attrition', df)
mod_1 = sm.OLS(outcome_1, predictors_1)
res_1 = mod_1.fit()
print(res_1.summary())
```

#### Description: 
Based on the information above, we found that there is a negative correlation between age and employee attrition, which means older employees are less likely to attrition (The coefficient term is -3.9536). A small p (≤ 0.05), reject the null hypothesis. This is strong evidence that the null hypothesis is invalid and there is a relation between age and attrition. Based on the p-value, age and attrition are statistically significant. Thus, we could conclude that there is a negative relationship between age and attrition. That is, for each additional age increase, the attrition would decrease by 3.9536.

### Contextual Factors:

#### Distance from employees' home to their office with Attrition

- Hypothesis Testion
  
  $H_{o}$: There is no relationship between distance from home and employee attrition
  
  $H_{A}$: There is relationship between distance from home and employee attrition

```python
outcome_1, predictors_1 = patsy.dmatrices('DistanceFromHome ~Attrition', df)
mod_1 = sm.OLS(outcome_1, predictors_1)
res_1 = mod_1.fit()
print(res_1.summary())
```

#### Description:
Based on the information above, we found that there is a positive correlation between the employees' home distance from the office and employee attrition, which means that when the distance of employees is longer, they are more likely to attrition (The coefficient term is 1.7173). A small p (≤ 0.05), reject the null hypothesis. This is strong evidence that the null hypothesis is invalid and there is a relation between the distance from home and attrition. Based on the p-value, the distance and attrition are statistically significant so we could conclude there is a positive relationship between distance from employees' home to office and attrition. For each additional increase in employees' home distance from the office, the attrition would increase by 1.7173.

#### Description:
Based on the coefficient term, we found that there is a negative correlation between employees' years at the company and employee attrition, which means that the longer employees stay at the company, they are less likely to attrition. A small p (≤ 0.05), reject the null hypothesis. This is strong evidence that the null hypothesis is invalid and there is a relation between years at the company and attrition. Based on the p-value, employees' years at the company and attrition are statistically significant, so we could conclude that there is a negative relationship between employees' years at company and attrition. For each additional increase in employees' years at the company, the attrition would decrease by 2.2382.
