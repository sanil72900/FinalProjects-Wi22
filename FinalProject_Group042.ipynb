## Project Overview:

In this project our team utilized data visualizations, linear-regression analysis, and chi-squared testing in order to examine potential differences between current UCSD students' occupational plans, preferred work industries, and mode of work preferences (remote vs in-person) in comparison to public datasets provided by the United States Bureau of Labor Statistics organization. 

The results of our analyses showed no correlation between student occupational plans and 2030 BLS Employment Projection rates. It turns out that current students are reporting their intent on finding work within the occupations of software development, public health, and tutoring/teaching at rates significantly higher than the 2030 Employment Projection rates, in which interest in software development work nearly doubled on the basis of student sampling (see fig. 1-4).

Concerning mode of work preferences per industry, it turns out that students are interested in working remotely within public administration, financial activity, and education and health services at rates higher than the BLS Telework Statistics dataset had predicted (see fig. 2-4).

Lastly, there is no difference between different UCSD graduating classes (2022-2025) concerning mode of work preference and, more importantly, students overall are are looking to work in-person rather than remotely, in which ~17% opted for remote work while ~34% opted for in-person work, given that their occupation would allow for the possibility of remote work (see fig. 3-5).

Our research question consists of three parts listed below:

#### I. Concerning employment trends over time:
Is there a correlation between the occupational plans of UCSD undergraduate students and the 2020-2030 US Bureau of Labor Statistics (BLS) Employment Projections? 

#### II. Concerning remote-work patterns:
Within the 2019 BLS Telework Statistics, there is a gap between reported ability to telework rates and actual takeup rates for persons who hold a bachelor's degree at minimum. How does this gap compare to the gap between student ability to telework rates and their interest in teleworking rates? How, if at all, do gaps vary across industries per dataset?

#### III. Concerning graduation status and interest in working remotely:
Do students within different graduation classes (2022-2025) show different levels of interest in working remotely after graduation?

Fig 1-4: Based on the scatterplot above, there seems to be very little to no correlation between UCSD students' occupational plans and the BLS 2030 Employment Projections.

In order to quantitatively detemine whether there truly is no correlation between UCSD students' occupational plans and the BLS 2030 Employment Projections, we will conduct a stastical analysis using a linear regression model.

```python
# Linear Regression
df4 = pd.DataFrame(df2, columns=['Occupation', 'Employment 2030 Percentage', 'UCSD Student Survey Percentage'])
df4 = df4.rename(columns={'Employment 2030 Percentage': 'Employment_2030_Percentage' , 'UCSD Student Survey Percentage': 'UCSD_Student_Survey_Percentage'})

outcome_1, predictors_1 = patsy.dmatrices('UCSD_Student_Survey_Percentage ~ Employment_2030_Percentage', df4)
mod_1 = sm.OLS(outcome_1, predictors_1)
res_1 = mod_1.fit()
print(res_1.summary())
```

On the basis of our regression model, we can see that the p-value is 0.011. The p-value is greater than the alpha value of 0.01. Therefore, we do not reject the null hypothesis which means that there is no positive correlation between UCSD students' occupational plans and the BLS 2030 Employment Projections.

```python
# Defining a function that converts all 'Remote-Preference' values to integers
def pref_convert(string):
    string=string.strip()
    string=string.lower()
    if string=='remotely':
        return 1
    else:
        return 0

# Defining a function that converts all 'Class level' values to integers
def classlevel_convert(string):
    string=string.strip()
    string=string.lower()
    if string=='freshman':
        return 1
    elif string=='sophomore':
        return 2
    elif string=='junior':
        return 3
    else:
        return 4

# Defining a function that converts all 'Gender' values to integers
def gender_convert(string):
    string=string.strip()
    string=string.lower()
    if string=='Male':
        return 1
    elif string=='Female':
        return 2
    elif string=='Non-binary':
        return 3
    else:
        return 4
```

```python
# Making a new DataFrame for chi-squared analysis
chi_df = survey[['Remote-Preference', 'Class level']]
chi_df['Remote-Preference'] = chi_df['Remote-Preference'].apply(pref_convert)
chi_df['Class level'] = chi_df['Class level'].apply(classlevel_convert)

# Chi-squared analysis and the results (for class level)
chi2_float, p_value, dof, expected = chi2_contingency(chi_df) 
 
print('Degrees of freedom is:', dof)
 
sig = 0.05

print("p-value is: " + str(p_value)) 

if p_value <= sig: 
    print('Reject null hypothesis') 
else: 
    print('Based on significant level = 0.05, accept null hypothesis')
```

The above code and output shows the results of chi-squared analysis of students' preferred mode of work across four different class levels (freshman, sophomore, junior, senior).
