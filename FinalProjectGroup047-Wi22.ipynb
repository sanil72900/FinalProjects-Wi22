# Research Question

Does the percent change of a country's IPAT value every 2 years between 2009-2018 show a strong correlation with Purchasing Power Parity (PPP) percent changes every 2 years for the same 10 year period?



*  An I=PAT value is a calculation made to measure the environmental impact of  humanity by using Population, GDP, and Greenhouse emissions data. We plan to calculate I=PAT by using Population data of each country, times GDP as measured by Purchasing Power Parity (PPP), times Greenhouse emissions by metric tons of carbon dioxide emitted (MtCO2e) that is divided by PPP. 
 * I = (Pop.) x (PPP) x (GHG/PPP).

## Background & Prior Work

As most are aware of today, the general warming of the earth as a result of human greenhouse gas emission carries far-reaching consequences for humanity. Given the vast variance in scales of economies around the world, it is clear that not every country is equally responsible for emissions. As we have observed, countries with a higher economic activity and output, indicated by their GDP, tend to be responsible for more carbon dioxide emissions to sustain this economic activity. This tends to be true both in the gross number and also per capita (3). One of the ways to measure the environmental impact is with an IPAT score. 

However, some research has also shown that percentage increases in GDP are correlated with decreases in carbon emissions, as richer nations develop more sustainable energy sources (1, 2). We examined the relationship between the PPP of countries and their emissions to see if we noticed any trends–indeed, many of the countries with the highest ratio were less-developed countries (4). Though they did not have the highest GDP, their overall emissions were far below the average. We wanted to further investigate whether there exists a significant correlation between these variables in two-year scales–that is, whether the fluctuations in countries’ PPP will be matched by similar movements in emissions. While the prior work we examined focused on this relationship for specific years, we want to examine changes over a multi-year time scale. Furthermore, it does not seem reasonable to separate GDP growth from environmental impact, where ultimately, GDP growth is unsustainable (5). 

Our project aims to explore the relationship between GDP growth and environmental impact using the aforementioned IPAT value. In this project we chose to use PPP to represent GDP because PPP takes into account factors such as cost of goods and services and inflation, which show a more balanced and representative measurement of quality of life compared to just looking at GDP per capita. To address confounding factors among different countries (such as country size), we will calculate the percentage change for the past 2 years of the country's IPAT score. Then, we will also get the percent change for the past 2 years of the country's PPP growth.

## Data Analysis -- Relationship Between Variables
Now we want to know if there's a correlation between the two variables GHG and GDP. So we calculated the Pearson value.

```python
# What is the relationship change in GHG emissions and change GDP between each set of years?

pearson_values = []

for yearset in ['0910', '1112', '1314', '1516', '1718']:
  outcome, predictors = patsy.dmatrices(f"ghg_{yearset} ~ gdp_{yearset}", clean_df)
  model = sm.OLS(outcome, predictors)
  ## fit the model
  results = model.fit()
  ## look at the results
  year_string = f"20{yearset[:2]} - 20{yearset[2:]}" 
  print(f"R-squared {year_string}: {results.rsquared}")
  pearson = np.corrcoef(clean_df[f"gdp_{yearset}"], clean_df[f"ghg_{yearset}"])[1][0]
  print(f"Pearson correlation {year_string}: {pearson}")
  print("\n")

  pearson_values.append(pearson)

pearson_df = pd.DataFrame({
    'Pearson Value': pearson_values, 
     'Year': ['2009-2010', '2011-2012', '2013-2014', '2015-2016', '2017-2018']
     })

sns.lineplot(data=pearson_df, x="Year", y="Pearson Value")
plt.suptitle("Correlation of GDP vs GHG (Pearson Value) Over Time")
```

# Machine Learning

Here we wrote a function to train our model and generate the plots. We are using a linear regression function where the dependent value is the GDP, and independent value is the IPAT or GHG. We are using the percent change for two 2-year intervals(15-16, 17-18).

```python
def model_data(X,y):
  Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,random_state=1)

  model = LinearRegression(fit_intercept=True)
  model.fit(Xtrain,ytrain)
  
  print('Training R^2',model.score(Xtrain,ytrain))
  print('Testing R^2',model.score(Xtest,ytest))
  
  # the predicted y values
  y_model = model.predict(Xtest)

  # create subplots
  fig,axes = plt.subplots(Xtrain.shape[1],1,figsize=[8,12])

  # if we are dealing with just 1 column
  if Xtrain.shape[1] == 1:
    sns.scatterplot(x = Xtest[Xtest.columns[0]],y = ytest,ax=axes)
    sns.lineplot(x = Xtest[Xtest.columns[0]],y = y_model,ax=axes)
    return
  
  coef_string = ' ' 

  # if we have multiple columns
  for i in range(Xtrain.shape[1]):
    sns.scatterplot(x = Xtest[Xtest.columns[i]],y = ytest,ax=axes[i])
    sns.lineplot(x = Xtest[Xtest.columns[i]],y = y_model,ax=axes[i])

    coef_string =  coef_string + ' ' +str(np.round(model.coef_[i],3))+ '*'+Xtest.columns[i]+ ' + '

  print('y = ',coef_string,' ',np.round(model.intercept_,3))
```

We write a function to remove outliers that are 2 standard deviations away from the mean. This function takes in a series, and uses the IQR (Interquartile range) and returns the series without outliers.

```python
def remove_outliers(df):

  """
  this function will remove outliers. Takes in a dataframe. Uses the IQR method and returns the dataframe without outliers.
  input: df
  output: df w/o outliers
  """
  meanval = np.mean(df)
  stdval = np.std(df)
  upper = meanval + stdval * 2
  lower = meanval - stdval * 2
  # print(meanval)
  # print(stdval)
  # print(upper)
  # print(lower)
  return np.where((df > lower) & (df < upper), df, np.nan)
```

Apply the function to remove outliers and use a heatmap to check for the outliers removed.

```python
cleaner_df = clean_df[['gdp_1718', 'ipat_1516', 'ipat_1718']]

for column in cleaner_df.columns:
  cleaner_df[column] = remove_outliers(cleaner_df[column])
```

# Ethics and Privacy

With our research question, we seek to find differences between CO2 emissions based on what their GDP and populations are. These measurements are all subject to their own biases. With the three different datasets, CO2, GDP, Population, we have to assume all the measurements are just estimates and not the exact value, and assume the possibility of errors in reporting data. We ultimately wanted to compare how different countries are contributing to climate change by their CO2 emissions, and see how reliant these countries' economic growth are to their emissions. A measure for exploring this topic is the IPAT score, which is: P for population, A for affluence, and T for technology. A measure in this equation that is most prone to bias technology as it is based on the viewpoint of what is causing the harm. For us this will be just CO2 emissions, but this value could be anything that could be deemed harmful. To get a better idea of the trend if countries are attempting to be less reliant on CO2 emissions based on PPP we will be using change in percentage in a two year period to more focus on how in the past 10 years countries have been changing. This should steer us from not just looking at pure IPAT scores as we are more interested in the trends countries are taking.

When looking at our data, GDP might not be the best measurement to summarize economic progress. That is why we will also use GDP per capita at purchasing power parity (PPP) to gain an insight into how economically well the people are living in the country. The CO2 and Population data are publicly available data posted on world banks. However, issues arise in how these measurements were collected as it’s usually not described, so they will serve better as estimates that may or may not be accurate.

Risk in our research project is minimal since the data is publicly available and doesn’t use personally identifiable information. It is unknown if the data we used was collected in a privacy violating manner. Still, nations usually agree to some sort of survey of their economic, population, and emissions contributions. While we hope most countries are honest about their data, we assume that there are data reliability checks before the data is released.

Evaluating our results from the data does not aim to target any country and serves as exploratory research comparing how a government is doing economically and looking at its contribution to environmental emissions. While our study seeks to find trends or at least a part of it, we aim to mitigate the effects of spurious variables by thoroughly using different tests and properly using machine learning algorithms to check the reliability of this method.

As more data becomes available, the exact or further enhanced models can test this method again. As of now, the data that is available from 2018 which is four years old data at the time of writing. If more recent data or better emissions metrics come available, our hypothesis should be tested again to verify its credibility and effectiveness.
