We start processing the data by filtering out the jobs that require more than just entry-level experience, as we're doing analysis on the basis that we're a newly-graduated data science student looking for the best place to start trailblazing towards the American Dream:

```python
def check_expertRequired(get_title):

    # basic string pre-processing
    get_title = get_title.lower()
    get_title = get_title.strip()

    # keyword filter
    exp = ['manager', 'senior', 'sr.', 'sr', 'lead', 'expert', 'experienced', 'principal', 'junior', 'jr.', 'jr']

    # filter for executive positions, and mark them if found
    for i in exp:
        if i in get_title:
            return 1
    else:
        return 0
```

```python
# create a new column and see if entries are correctly flagged
ds_jobs['expertRequired'] = ds_jobs['Job Title'].apply(check_expertRequired)
ds_jobs.head()
```

```python
# remove flagged entries, and check that all jobs
# above entry level positions have been removed
ds_jobs = ds_jobs[ds_jobs.expertRequired == 0]
ds_jobs.head()
```

We also remove superfluous columns:

```python
ds_jobs = ds_jobs.drop(['Unnamed: 0', 'index',
       'Job Description', 'Rating', 'Headquarters',
       'Size', 'Founded', 'Type of ownership', 'Industry', 'Sector', 'Revenue',
       'Competitors', 'Easy Apply', 'expertRequired'], axis = 1)
```

```python
ds_jobs.head()
```

We need to be able to sort and otherwise analyze job posting salaries as
usable numbers, so we standardize the estimated Glassdoor salary range distances as an integer range:

```python
# salary standardization function
def standardize_distance(salary):
      
    # basic string pre-processing
    salary = salary.lower()
    salary = salary.strip()
    
    # strip the salary estimate to a raw integer distance range
    salary = salary.replace("k", "000")
    salary = salary.replace("(glassdoor est.)", "")
    salary = salary.replace("$", "")
    salary = salary.replace("(employer est.)", "")
    salary = salary.strip()
    if (salary == '34-53 per hour') | (salary == '10-26 per hour') | (salary == '17-27 per hour') | (salary == '-1'):
        salary = np.nan
    
    return salary
```

```python
# standardize the estimate column
ds_jobs["Salary Estimate"] = ds_jobs["Salary Estimate"].apply(standardize_distance)
ds_jobs.head(3)
```

We also simplify column names for future analysis and readability.

```python
ds_jobs = ds_jobs.rename(columns={'Job Title': 'job','Salary Estimate':'salary', 'Company Name': 'company', 'Location': 'location'})
```

```python
ds_jobs.head()
```

### Union of Data Scientist and Data Analyst datasets
We want to make one master list to account for how fresh Data Science graduates commonly enter the industry though Data Scientist and Data Analyst positions.<br>
<br>
However, since the Data Scientist and Data Analyst job posting datasets were both scraped from Glassdoor, we reckoned there would surely be overlapping postings.<br>
<br>
We will resolve this by merging the two datasets and removing duplicate entires by filtering out overlapping job positions in the same company.

```python
# remove any overlapping job postions in the same company
tmp = da_jobs[da_jobs["company"].isin(ds_jobs["company"])]
ds_jobs = ds_jobs[~ds_jobs["job"].isin(tmp["job"])]

# merge two datasets 
ds_jobs = pd.concat([ds_jobs, da_jobs], ignore_index=True)

# drop the company column
ds_jobs = ds_jobs.drop(['company'], axis = 1)

ds_jobs.head()
```

We split the (lower, upper) salary range and (city, state) location columns into two parts for easier analysis down the road.

```python
# split location into two parts - city and state
new = ds_jobs["location"].str.split(",", n = 1, expand = True)
ds_jobs["city"]= new[0]
ds_jobs["state"]= new[1]
ds_jobs.drop(columns =["location"], inplace = True)

# split salary into two parts - predicted highest salary and lowest salary
new_2 = ds_jobs["salary"].str.split("-", n = 1, expand = True)
ds_jobs["low salary"] = pd.to_numeric(new_2[0])
ds_jobs["high salary"] = pd.to_numeric(new_2[1])
ds_jobs.drop(columns =["salary"], inplace = True)
```

```python
# https://stackoverflow.com/questions/13636848/is-it-possible-to-do-fuzzy-match-merge-with-python-pandas
# define a merging function
 
import difflib 

def fuzzy_merge(df1, df2, left_on, right_on, how='inner', cutoff=0.6):
    df_other= df2.copy()
    df_other[left_on] = [get_closest_match(x, df1[left_on], cutoff) 
                         for x in df_other[right_on]]
    return df1.merge(df_other, on=left_on, how=how)

def get_closest_match(x, other, cutoff):
    matches = difflib.get_close_matches(x, other, cutoff=cutoff)
    return matches[0] if matches else None
```

```python
# merge aggregated average salaries and cost of living index
mrg = fuzzy_merge(city_salary_avg,
                  city_cola,
                  left_on = 'city',
                  right_on = 'index')
```

Now, having seen the aggregate average salaries and expenses, how long would it take to purchase a single-detached home in each of these cities?
<br><br>
With some math, we can calculate an estimate of how long it would take to purchase a single-detached house in each city. We divide the house price by the result of the average Data Science entry level job salary minus our estimated cost of living.
<br><br>
We have our formula:
<br><br>
$Duration\ to\ Purchase\ a\ Single\ Detached\ House = \frac{the\ Single\ Detached\ House\ Price}{Average\ DS\ Salary\ for\ a\ City - the\ Cost\ of\ Living\ for\ a\ City}$
<br><br>
We'll calculate this and store it in  new column in our merged dataframe.

```python
mrg['div'] = mrg['house price'] / mrg['possible savings a year']
```

Finally, we return to our our cities of interest.<br>
We'll calculate their final rankings to answer our question of, "what is the best city for a UCSD Data Science student who wants to purchase a single-family detached house?"<br>
The final rankings will be calculated from a rank sum of each feature of:<br>
- Calculated duration to purchase a house
- Lowest aggregate average housing prices
- Highest aggregate average salary (after average annual expenditures)
- Cost of living

```python
# create df_rank, add duration column with inner merge with mrg
df_rank = df.drop(['min_salary', 'max_salary'], axis=1)
df_rank = pd.merge(df_rank, mrg[['index', 'div']], how='inner', on='index')

# convert columns to rankings
df_rank['prices'] = df['prices'].rank(ascending=False)
df_rank['mean_salary'] = df['mean_salary'].rank()
df_rank['cost_of_living'] = df['cost_of_living'].rank(ascending=False)
df_rank['div'] = df_rank['div'].rank(ascending=False)

# create rank column and sort by rank
df_rank['rank'] = df_rank.sum(axis=1)
df_rank = df_rank.sort_values(by='rank', ascending = False)
df_rank.columns = ['City Name', 'House Price', 'Mean Salary', 'Cost of Living', 'Duration', 'Rank']
df_rank
```
