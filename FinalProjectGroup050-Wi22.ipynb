#### [3][1] San Diego Traffic Dataset

- [0] This database includes data on the number of vehicles on an average weekday at select street intersections in San Diego.

```python
### read in traffic dataset
traffic = pd.read_csv('https://seshat.datasd.org/traffic_counts/traffic_counts_datasd_v1.csv')
```

- [1] Now, we need to remove non-relevant data from the dataset
- [2] Then, extract the year from the date for future analysis

```python
### remove unnecessary columns
traffic = traffic[['street_name','limits','total_count','date_count']]

### extract year from date_count
traffic['year'] = pd.to_datetime(traffic['date_count'])
traffic['year'] = traffic['year'].dt.year
```

- [3] Next, we standardize our street name, so google Maps' geocoding API will recognize the street name
- [4] We also combine each street's boundary limit with the street name to quickly compute latitude and longitude later on

```python
### standardize street name and concatenate limits into street name
street_name = []
for limits,str_name in zip(traffic.limits, traffic.street_name): 
    
    try:
    
        if str_name == '01 AV':
            str_name = 'FIRST AV'
        elif 'RH' in str_name:
            str_name = str_name.replace('RH', 'RANCH')
        else:
            str_name = str_name
        
        if '-' in limits:
            limit = limits.split('-')
            if len(limit) >= 2:
                limit = limit[len(limit)-2].strip()
                street_name.append(str_name + ' & ' + limit + ', San Diego, CA')     
        else:
            street_name.append(str_name)
    except:
        street_name.append(str_name)
```

- [5] After standardizing our street names, we now replace the un-kept street data with a cleaned version and drop observations with all missing values

```python
### replace street name with standardized version and remove used data
traffic['street_name'] = street_name
traffic.drop(['limits'], axis=1, inplace=True)

### remove rows where all of the columns have Null value
traffic.dropna(axis='index', how='all', inplace=True)
```

#### --> [3][5] Calculate Traffic within Certain Radius of Each Stop

- [0] Now we have computing location data for both the blue line stops dataset and traffic dataset, we can calculate traffic around each stop

```python
### given radius and traffic data, calculate traffic around each stop
### return a list of traffic counts of each stop
def cal_traffic(r,traffic_data):
    stop_traffic = []
    for i_b, row in blue_line.iterrows():
        traffic_count = 0
        for i_t, obs in traffic_data.iterrows():
            if obs['lat'] !=0 and obs['lon'] !=0:
                try:
                    dist = geodesic((row['stop_lat'],row['stop_lon']),(obs['lat'],obs['lon'])).miles
                    if dist <= r:
                        traffic_count = traffic_count + obs['total_count']
                except:
                    pass
        stop_traffic.append(traffic_count)
    return stop_traffic 
```

- [1] We chose to use a radius of 2 miles for our geolocation data, as after testing values between 1 and 3 miles between stops, we found the value of 2 miles to return the most consistent data for our further analysis. 

```python
### set radius(mile)
radius = 2

### calculates traffic of each stop of each year in the traffic data
bl_col = sorted(traffic_result['year'].value_counts().index)
for i in bl_col:
    blue_line[str(i)] = cal_traffic(radius,traffic_result[traffic_result['year']==i])
```

When excluding the years containing outliers in the dataset pertaining to blue line stops surrounding the UTC area, our data can generate a line of best fit, where a similar upwards trend can be observed compared to the scatterplot developed for the downtown areas. We believe that this also retains particular relevance to the aforementioned housing crisis referenced in section 1.2, as the exceedingly expensive real estate market surrounding the UCSD campus leads to an increased need for students to have ample access to the campus from further housing locations.

A significant caveat to our analysis and conclusion that cannot be ignored is the Covid-19 pandemic, which significantly (or almost entirely) negated the need for public transit during lockdown periods and especially remote instruction from UCSD. Fortunately, since we decided to drop years from our datasets that the pandemic would influence, this didn't pose any significant issues for us as a group regarding data analysis or visualization.
