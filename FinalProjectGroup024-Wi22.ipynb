# Overview

The SAT has served as a major factor in high school students’ college application process. Over recent years, there have been various controversies on whether or not a student’s socio-economic background acts as a major factor in his or her performance on the exam. Critics of the exam argue that the SAT is biased against students of lower privileged demographics who are unable to access the same resources as their higher privileged peers. The purpose of our project was to precisely determine whether the timeframe of COVID-19 distance learning disproportionately affected Californian minority students’ SAT scores. This project drew on official data obtained from The College Board from the years of 2017-2021, which included score breakdowns based on student ethnic groups. We focused specifically on students residing in California in order to simplify our analysis. Based on the data we visualized using the pandas, seaborn, and matplotlib libraries, we did in fact come to the conclusion that the timeframe of COVID-19 distance learning impacted students of certain minority groups more than others.

## Background & Prior Work

With the COVID-19 pandemic globally affecting the social, economic, and political fields around the world, it has also affected the way educational institutions facilitate standardized testing, in particular the SAT. Following the news of COVID-19 amid the school year, education drifted toward a more online-based structure through Zoom, Canvas, etc. This entails online testing. Rather than the standard in-person exams side to side with their fellow peers, the students are restricted to a more isolated zone in front of their monitors. With distractions back to back and social isolation, the new environment has been devastating for a multitude of students. In response to this change in learning, the College Board initially utilized an online testing platform for students from 2019 to 2020. Within the following years, they implemented limited in-person testing centers to simulate the traditional standardized testing. However, despite their efforts to continue standardized testing, a multitude of colleges began to modify their application standards to make SAT testing optional or even obsolete [1]. With the drastically different learning environments and the lack of demand for standardized testing, the progression of SAT scores has been up for question. While the correlation between household income and SAT scores has been proven, we want to put this correlation in context with the COVID-19 pandemic [4], in particular focusing on the comparison between minorities and non-minorities. In our project, we will identify whether COVID-19 has made a drastic change in the SAT scores throughout the years before and after the COVID-19 pandemic.

In October of 2020, Scott Jaschik of Inside Higher Ed [2] exclaimed the average SAT score between 2019 and 2020 decreased from 1059 to 1051. College Board’s data indicated that 83% of Asian students met the benchmark for reading and writing, with 80% for the mathematics benchmark. In contrast, only 44% of black students met the benchmark for reading and writing and 21% for mathematics.
In September 2021, Jaschik [3] noted that 1.5 million students in the class of 2021 took the SAT at least once, 700,000 lower than the class of 2020. In regards to student performance, the average SAT score increased for the class of 2021 (1051 to 1060). However, certain patterns of previous years remained consistent, such as performance-based on race (with Asian Americans scoring better on average). Jaschik’s observations beg the question of whether the advent of the COVID-19 pandemic augments or undermines the already existing discrepancy between the SAT scores of students of different socioeconomic backgrounds.

# Hypothesis


We hypothesize that California minority students, who are disproportionately affected by COVID-19, will have a greater decrease in their SAT scores from before to after COVID-19 compared to students that are not minorities. With an unfair disadvantage in educational resources and online learning, minority students would potentially encounter more difficulties in studying for and taking the SATs. For those students who are non-minority and perhaps less affected by COVID, we expect little change in their average scores.

### Proportional Change of Test Takers

```python
#picking subset of data
test_takers = df[['year', 'race','num_test_takers']]

#creating new dataframes for each race with a new column containing percent change of test takers from year to year
data1 = test_takers[test_takers['race'] == 'American Indian/Alaska Native']
data1_pc = data1['num_test_takers'].pct_change().to_frame()

data2 = test_takers[test_takers['race'] == 'Asian']
data2_pc = data2['num_test_takers'].pct_change().to_frame()

data3 = test_takers[test_takers['race'] == 'Black/African American']
data3_pc = data3['num_test_takers'].pct_change().to_frame()

data4 = test_takers[test_takers['race'] == 'Hispanic/Latino Test Takers']
data4_pc = data4['num_test_takers'].pct_change().to_frame()

data5 = test_takers[test_takers['race'] == 'Native Hawaiian/Other Pacific Islander']
data5_pc = data5['num_test_takers'].pct_change().to_frame()

data6 = test_takers[test_takers['race'] == 'White']
data6_pc = data6['num_test_takers'].pct_change().to_frame()

data7 = test_takers[test_takers['race'] == 'Two or More Races']
data7_pc = data7['num_test_takers'].pct_change().to_frame()
```

```python
#adding original df and df containing percent changes to respective lists
data_list = [data1, data2, data3, data4, data5, data6, data7]
data_pc_list = [data1_pc, data2_pc, data3_pc, data4_pc, data5_pc, data6_pc, data7_pc]

#iterate through all the dataframes, adding a column of year ranges and a column of percent changes
for i in range(len(data_list)):
    data_list[i]['year span'] = ['NaN','2017 - 2018', '2018 - 2019', '2019 - 2020','2020 - 2021']
    data_list[i]['percent test takers'] = data_pc_list[i]['num_test_takers']

#concat the edited dataframes into one dataframe, dropping null values and extraneous columns
sat_pc = pd.concat(data_list).dropna().set_index('year span').drop(['year', 'num_test_takers'],axis=1)
sat_pc.head()
```

```python
fig, axes = plt.subplots(1, 6, figsize=(40, 10))
sat_pc[sat_pc['race']=='American Indian/Alaska Native']['percent test takers'].plot(kind='bar', ax=axes[0], title='% Change in American Indian/Alaska Native Test Takers', legend=False)
sat_pc[sat_pc['race']=='Asian']['percent test takers'].plot(kind='bar', ax=axes[1], title='% Change in Asian Test Takers', legend=False)
sat_pc[sat_pc['race']=='Black/African American']['percent test takers'].plot(kind='bar', ax=axes[2], title='% Change in Black/African American Test Takers', legend=False)
sat_pc[sat_pc['race']=='Native Hawaiian/Other Pacific Islander']['percent test takers'].plot(kind='bar', ax=axes[3], title='% Change in Native Hawaiian/Other Pacific Islander Test Takers', legend=False)
sat_pc[sat_pc['race']=='White']['percent test takers'].plot(kind='bar', ax=axes[4], title='% Change in White Test Takers', legend=False)
sat_pc[sat_pc['race']=='Two or More Races']['percent test takers'].plot(kind='bar', ax=axes[5], title='% Change in 2+ Races Test Takers', legend=False)
```

Before 2020, there has been a slight increase in test takers overall, however each racial group has a different percent increase in test takers. The number of American Indian/Alaska Natives test takers have seen a dip in test takers in 2019. The number of Asian test takers have stagnated from 2018 to 2020. Black/African American, Hispanic/Latino, and students of mixed race have increased slightly. White and Native Hawaiian/Other Pacific Islander test takers have decreased from 2018 to 2020. 

After 2020, there has been a decrease in the number of test takers for each racial group, however the percent decrease is different among the racial groups. White, Native Hawaiian/Other Pacific Islander, Hispanic/Latino, and Black/African American students have seen a similar percent decrease of 60 percent from 2020 to 2021. Students of mixed race have also seen a decrease of 60 percent. On the other hand, Asian and American Indian/Alaska Native students have seen a decrease of 40 percent from 2020 to 2021. 

### Statistical Significance

Since we have the most data on Hispanic/Latino and White students, let's see if there is a statistically significant difference in their scores using a t-test.

```python
#Tells us if there is a significance between Hispanic/Latino and White students' SAT total score
hl_score = data[data['race'] == 'Hispanic/Latino']['total_score']
w_score = data[data['race'] == 'White']['total_score']
t_val, p_val = ttest_ind(a = hl_score, b = w_score)

# Check if statistical test passes significance, using an alpha value of 0.01. This code provided.
if p_val < 0.01:
    print('Data Science accomplished, there is a significant difference!')
else:
    print('There is NOT a significant difference!')
```
