## Background & Prior Work

As weather drops to severely low temperatures, the football itself contracts, noticeably affecting gripping, throwing, and kicking (2). However, in freezing temperatures, passing accuracy only decreases by 2%, field goal accuracy decreases by 2%, and average punting distance decreases by 3 yards (2). Further, NFL games from 1994 to 2016 averaged 42.1 points when temperatures were above freezing and below 80 degrees Fahrenheit, 42.6 points when below 32 degrees, 42.8 points when below 20 degrees, and 43.3 points when below 10 degrees (2). These findings suggest that scoring and football players' performance is not affected by severe weather conditions, but does not address whether different weather conditions impacts the offensive plays being called.
Pro Football Focus also reports that quarterbacks suffer a 1.8% decline in their completion percentage and a 0.17-year decline in their yards per attempt in cold-weather games (3). Quarterback performance is further affected by below-freezing weather conditions, shown by a 3.1% decline in completion percentage and 0.19-yard decline in their yards per attempt in these conditions (3). This results in a 5-7% decrease in passing rates during snow games (3). While this article argues that weather is a relevant and significant performance factor, it does not attribute the ability of football teams to still succeed in severe weather to play-calling. Therefore, it would be interesting to analyze whether severe weather drastically impacts offensive plays being made during NFL games.

# Hypothesis


We hypothesize that if games are played in harsh weather conditions such as freezing temperatures, high winds (20mph+), and NFL rainfall levels defined as heavy, teams will play more conservatively (less passing plays and more rushing). We think this because of the fact that it becomes more difficult to complete passes when the ball is wet/cold, the field is difficult to run on, players' vision is impaired, or when the winds, snow, or rain are severe.

# Part 1

We wanted to start with exploring the number of plays called per game in relation to the different weather conditions.

```python
#here we are replacing all "Sack" values with "Pass" because if the QB was sacked this means a pass play was called
clear_play_types['PlayType'].replace({'SACK': 'PASS'}, inplace = True)
moderate_rain_play_types['PlayType'].replace({'SACK': 'PASS'}, inplace = True)
heavy_rain_play_types['PlayType'].replace({'SACK': 'PASS'}, inplace = True)
snow_play_types['PlayType'].replace({'SACK': 'PASS'}, inplace = True)
heavy_wind_play_types['PlayType'].replace({'SACK': 'PASS'}, inplace = True)
```

```python
#here we are getting all the value counts of each playtype to help make the graphing process smoother
clear_counts = clear_play_types['PlayType'].value_counts().to_list()
moderate_rain_counts = moderate_rain_play_types['PlayType'].value_counts().to_list()
heavy_rain_counts = heavy_rain_play_types['PlayType'].value_counts().to_list()
snow_counts = snow_play_types['PlayType'].value_counts().to_list()
heavy_wind_counts = heavy_wind_play_types['PlayType'].value_counts().to_list()
```

```python
#grouping by GameId and getting the number of rows so we have the total number of games played in each condition
clear_game_count = clear_play_types.groupby('GameId').count().shape[0]
moderate_rain_game_count = moderate_rain_play_types.groupby('GameId').count().shape[0]
heavy_rain_game_count = heavy_rain_play_types.groupby('GameId').count().shape[0]
snow_game_count = snow_play_types.groupby('GameId').count().shape[0]
heavy_wind_game_count = heavy_wind_play_types.groupby('GameId').count().shape[0]
```

```python
#here we are looping through all the play counts and dividing by the number of total games played in each weather 
# condition so we can normalize the y axis of our barplots to show the number of each play type called per game
for i in range(len(clear_counts)):
    clear_counts[i] = clear_counts[i] / clear_game_count
    
for i in range(len(moderate_rain_counts)):
    moderate_rain_counts[i] = moderate_rain_counts[i] / moderate_rain_game_count
    
for i in range(len(heavy_rain_counts)):
    heavy_rain_counts[i] = heavy_rain_counts[i] / heavy_rain_game_count
    
for i in range(len(snow_counts)):
    snow_counts[i] = snow_counts[i] / snow_game_count
    
for i in range(len(heavy_wind_counts)):
    heavy_wind_counts[i] = heavy_wind_counts[i] / heavy_wind_game_count
```

```python
# rounding the play counts per game to one decimal
clear_counts = np.around(clear_counts, 1)
moderate_rain_counts = np.around(moderate_rain_counts, 1)
heavy_rain_counts = np.around(heavy_rain_counts, 1)
snow_counts = np.around(snow_counts, 1)
heavy_wind_counts = np.around(heavy_wind_counts, 1)
```

```python
#Creating barplots of the number of plays by play type called per game in different levels of rain
fig, (ax1, ax2) = plt.subplots(ncols = 2, sharey = True)
heavy_rain = sns.barplot(data = heavy_rain_play_types, x = heavy_rain_counts, y = heavy_rain_counts, ax = ax1, palette = 'crest')
moderate_rain = sns.barplot(data = moderate_rain_play_types, x = moderate_rain_counts, y = moderate_rain_counts, ax = ax2, palette = 'crest')
ax1.set_title('Heavy Rain Plays Per Game')
ax2.set_title('Moderate Rain Plays Per Game')
ax1.set_xlabel(' FGs         Punt        Rush        Pass')
ax2.set_xlabel(' FGs         Punt        Rush        Pass')
```

We can see from this plot that the number of plays called in games with heavy rain is quite a bit less than the number of plays in games with moderate levels of rain. To deal with this we can find the percent of each play type called and compare them between the different rain levels.

```python
#here we are getting the percent of each play type that gets called
pass_plays_hr_percent = (heavy_rain_counts[0] / heavy_rain_counts.sum()) * 100
pass_plays_mr_percent = (moderate_rain_counts[0] / moderate_rain_counts.sum()) * 100
rush_plays_hr_percent = (heavy_rain_counts[1] / heavy_rain_counts.sum()) * 100
rush_plays_mr_percent = (moderate_rain_counts[1] / moderate_rain_counts.sum()) * 100
punt_plays_hr_percent = (heavy_rain_counts[2] / heavy_rain_counts.sum()) * 100
punt_plays_mr_percent = (moderate_rain_counts[2] / moderate_rain_counts.sum()) * 100
fg_plays_hr_percent = (heavy_rain_counts[3] / heavy_rain_counts.sum()) * 100
fg_plays_mr_percent = (moderate_rain_counts[3] / moderate_rain_counts.sum()) * 100
[(pass_plays_hr_percent, pass_plays_mr_percent), (rush_plays_hr_percent, rush_plays_mr_percent), 
 (punt_plays_hr_percent, punt_plays_mr_percent), (fg_plays_hr_percent, fg_plays_mr_percent)]
```

What we can see from this quick analysis is that in different types of rainy conditions there is almost no difference in the type of plays called. There is around 2% more passing plays called in heavy rain and 2% more rushing plays called in moderate rain but this is a very small difference. Now lets look at clear, snowy and windy weather conditions.

```python
#creating plots for snowy, windy, clear conditions
fig, (ax1, ax2, ax3) = plt.subplots(ncols = 3, sharey = True)
snowy = sns.barplot(data = snow_play_types, x = snow_counts, y = snow_counts, ax = ax1, palette = 'mako')
windy = sns.barplot(data = heavy_wind_play_types, x = heavy_wind_counts, y = heavy_wind_counts, ax = ax2, palette = 'rocket')
clear_conditions = sns.barplot(data = clear_play_types, x = clear_counts, y = clear_counts, ax = ax3, palette = 'viridis')

ax1.set_title('Snow Plays Per Game')
ax2.set_title('Wind Plays Per Game')
ax3.set_title('Clear Plays Per Game')
ax1.set_xlabel(' FGs  Punt  Rush  Pass')
ax2.set_xlabel(' FGs  Punt  Rush  Pass')
ax3.set_xlabel(' FGs  Punt  Rush  Pass')
```

```python
rush_data = pd.DataFrame(plays.groupby('GameId').sum()['IsRush']).reset_index()
pass_data = pd.DataFrame(plays.groupby('GameId').sum()['IsPass']).reset_index()
temperature_data = pd.DataFrame(weather.groupby('game_id').mean()['Temperature']).reset_index()
yards_df = pd.DataFrame(plays.groupby('GameId').sum()['Yards']).reset_index()
precipitation_data = weather.groupby('game_id').mean()['Precipitation'].reset_index()
compiled_data = rush_data.merge(temperature_data,left_on = 'GameId', right_on = 'game_id')
compiled_data = compiled_data.merge(pass_data, left_on = 'GameId', right_on = 'GameId')
compiled_data = compiled_data.merge(yards_df, left_on = 'GameId', right_on = 'GameId')
compiled_data = compiled_data.merge(precipitation_data, left_on = 'GameId', right_on = 'game_id')
```

```python
plays['turnover'] = plays['IsFumble'] + plays['IsInterception']
turnover_data = plays.groupby('GameId').sum()['turnover'].reset_index()
compiled_data = compiled_data.merge(turnover_data, left_on = 'GameId', right_on = 'GameId')
```

```python
points = []
for i in range(plays.shape[0]):
    if plays.iloc[i].IsTouchdown == 1:
        points.append(6)
    elif plays.iloc[i].PlayType == 'FIELD GOAL':
        points.append(3)
    elif plays.iloc[i].PlayType == 'EXTRA POINT':
        points.append(1)
    elif plays.iloc[i].PlayType == 'TWO-POINT CONVERSION':
        points.append(2)
    else:
        points.append(0)
plays['Total_Points'] = points
total_points_df = pd.DataFrame(plays.groupby('GameId').sum()['Total_Points']).reset_index()
compiled_data = compiled_data.merge(total_points_df, left_on = 'GameId', right_on = 'GameId')
```

```python
ToGo_data =  pd.DataFrame(plays.groupby('GameId').mean().get('ToGo')).reset_index()
final_compiled_data = compiled_data.merge(ToGo_data, left_on = 'GameId', right_on = 'GameId')
```

```python
compiled_data = final_compiled_data
compiled_data = compiled_data.drop(columns = ['game_id_x','game_id_y'])
```

# Conclusion & Discussion

The research question we investigated was how the severity of weather, particularly temperature, rainfall, and wind rate affect how aggressively a team might conduct their offense. In addition, we also explored how successful offenses are at scoring points and gaining yards in NFL games from 2013 to 2021. From our background research, we discovered that weather has been found to be a relevant and significant performance indicator for NFL teams. However, play-calling has thus far not been attributed to the ability of NFL teams to still succeed in severe weather. To analyze whether severe weather impacts offensive plays being called during NFL games, we used a weather dataset with information about temperature, precipitation, wind, etc. during NFL games from 2000-2021. We then matched the game ids provided from the weather dataset to the game ids provided by play-by-play datasets we found for NFL games from 2013-2021. In exploring the data during our initial data analysis, we plotted the number of plays called (field goals, punts, rush plays, and pass plays) by weather condition. In doing so, we found that while there seemed to be a significant difference between play-call types for heavy rain vs. moderate rain, the play-calls for snow, wind, and clear weather conditions seemed negligible. Thus, for the rest of our data analysis, we decided to focus on temperature and precipitation.



Neither the precipitation, nor the temperature recorded, seemed to show a statistically significant relationship with the turnover rate per game. However, there was a statistically significant relationship between temperature and total points scored and temperature, meaning that there are generally more points scored in warmer weather. In exploring pass plays specifically, we found that there was a significant positive correlation for temperature, meaning that as temperature increases, pass plays are more likely to be called. However, the amount of precipitation did not seem to significantly affect pass plays called. In contrast, the relationship between temperature and the amount of rush plays called was not found to be statistically significant. However, we did find a relatively strong and positive correlation between rush plays and precipitation, meaning that as rain becomes heavier, the amount of rush plays called within a game is also likely to increase. The total yards gained within a game was found to have a statistically significant and positive correlation with temperature, meaning that as the weather becomes warmer the total yards gained within a game is likely to increase. However, like the pass plays data, precipitation did not seem to significantly affect total yards gained. Finally, when exploring the ToGo data, temperature had a significantly positive correlation, implying that as temperature increases, the yards needed to get a new set of downs also increases. However, the precipitation did not significantly impact this measure. Overall, this data partially supports our hypothesis; while we found that snow, wind, and clear weather has minimal impact on play calls, unlike our prediction, we found that severe precipitation and temperature does in fact affect aggressive play calling.



In exploring this data, we decided to only focus on data given in a numerical format, or data which we could convert to a numerical format. Thus, we were limited to mainly exploring temperature and precipitation over other weather conditions because the range of data points was much greater, in comparison to data like snowfall which was only categorized into heavy, moderate, and light. Thus, while we can make conclusions about how temperature or precipitation affects NFL play-calling, we are unable to extend our findings to the other variables that undeniably affect weather. Even so, our findings may have a visible impact on society. For one, this data may affect the future of the NFL in that more and more stadiums may be built as domes. This essentially eliminates weather as a factor in the sport, and drastically reduces the home-field advantage that many teams have. In knowing this information, play-calling strategies may change significantly in the future as weather will essentially become a non-factor in NFL games.
