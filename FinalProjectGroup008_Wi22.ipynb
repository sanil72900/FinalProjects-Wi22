# Background

**Overview:**
In the highly volatile, early stages of a company's growth phase, retail investors can find it difficult to objectively identify the companies that will bring great returns with relatively limited risk. A small-cap company has a market cap between \$300M and $2B [2]. Small-cap companies have a larger growth potential because they control less market share, which means they present an opportunity for larger returns at the cost of higher risk. We were curious to see which financial metrics best indicated the company’s future success. Why were some successful while others failed?

**Distinguishing Dynamic of Small-Cap Companies:**
Small cap companies are difficult to invest in because they can’t be analyzed in the same way as larger companies. Many don’t have profits or have limited cash, but that doesn’t mean they are bad investments if you know what to look for. For smaller companies, we are more concerned with metrics that indicate rapid growth that can set the company up to be profitable in the future [9]. This is in contrast to investing in larger companies where investors look for steady, sustainable growth. However, the faster a company grows, the higher the risk for it to fail when they can't keep up with their leverage. Corporate management may take higher risks such as taking out large amounts of debt to leverage future growth. This can pay off to a certain extent, but there comes a point when such practices become irresposible. Investors must be aware of these risks and choose the companies where the potential benefits outweigh the risks, increasing the likelihood of a successful investment. In order to identify such long-term invest opportunities, one can use a process called fundamental analysis to determine a company’s intrinsic value and see if the company has good financial health [10].

**Initial Approach:**
We made our analysis following this guide: According to a Nasdaq investing guide [1], debt/equity ratio and cash flow are important metrics when analyzing smaller companies. Small cap stocks have the best performance when we omit those with low profitability [3]. We want to observe stocks in the 2014 to 2018 timeframe because this prevents the effects of the 2008 recession and the late-2018 pullback from distorting our analysis.

Fundamental analysis variables we want to use:
*   Revenue: the top-line figure for money generated from normal business operations [12]
*   Gross Margin: net sales minus cost of goods sold; what the company retains after direct costs of producing it's products/services [13]
*   Earnings Before Tax: the revenue minus expenses other than tax
*   Net Income: sales minus cost of goods sold [14]
*   Debt to Equity: the ratio of how much a company is financing it's operations versus wholly owned funds; measure of the ability for a company to cover all outstanding debts in case of a downturn [15]
*   Cash and Cash Equivalents: the value of a company's assets that are held in cash or can be converted to cash immediately; in other words, liquidity [16]
*   Total Liabilities: the total debts and obligaitons a company owes to outside parties [17]
*   Operating Cash Flow: a measure of the amount of cash generated by a company's normal business operations [18]
*   Free Cash Flow: the cash a company generates after accounting for cash outflows to support operations and maintain its capital assets [19]
*   R&D Expenses: expenses are associated directly with the research and development of a company's goods or services and any intellectual property generated in the process [20]

**Potential Complications:**
We anticipate there will be some confounders that could present a challenge to our analysis. The primary issue would be association between variables since some financial metrics are naturally linked. For example, high revenues are typically related with high income before tax. This would mean that errors aren't entirely random and rather influenced by some other factor. For example, tomorrows stock price is heavily infuenced by yesterday's stock price. This if we were trying to design a predicted for the price of an index in the future, we should use change in stock price to avoid association. In a similar manner, we plan to use change in our financial metrics over the 5 years to limit issues arising from association. Another confounder could be that analyzing a 5-year span introduces some natural variability in the markets that are influenced by factors external to the fundamentals of a company [11]. For example, even if \$AAPL is reporting consistent fundamentals, their stock price could be negatively impacted due to uncertainty in the market. Over the span of 2014 to 2018, the most significant of these factors would be the presidential election of 2016 and market pullback of 2018, which resulted from the Federal Reserve unexpectedly raising rates. Fear, uncertainty, and doubt (FUD) in the market leads investors to limit risk by selling equities and moving to less risk-prone securities such as bonds, cash, or precious metals. The level of FUD in the market can be estimated by the volatility index (\$VIX), where values over 25 generally indicate significantly high levels of volatility in the market; as expected, there are spiked in the \$VIX during the times corresponding to the election of late-2016 and Fed rate hike of late-2018 [8]. This selling pressure can weaken correlation between the fundamentals of a company and the stock price (which is directly linked to market cap). Thus, such variability in the market could weaken our ability to create a model that uses certain financial indicators 

**Previous work:**
Although no significant study has been done on this topic specifically, there is research on the prediction of a company’s failure through analysis of its financial ratios. The first study was done on a total of 140 companies(70 successful and 70 failed) in the Colombo stock market in Sri Lanka over the years 2002 to 2008 [4]. Using logistic regression analysis on 15 financial ratios, the study produced a model that was 77.86% accurate 1 year before the company’s failure. Another similar study was conducted on companies in the Istanbul Stock Exchange over the years 2009 to 2011 [5]. This research found that out of 20 available financial ratios, 5 of them were the most important in determining a company’s success. Some of these include capital adequacy, net working capital, and total asset ratios. The model for this study was around 90% accurate in predicting a company’s failure throughout those years.

# Hypothesis

We believe that a selection of a company’s financials will allow us to determine the success of small-cap companies in the US as measured by market capitalization growth over a mid-to-long-term period of 5 years. More specifically, growth in revenue and growth in cash and cash equivalents will carry the most significance in establishing a correlation to market cap growth. Revenue is a great indicator for how rapidly a company is able to grow because as this top-line figure increases, it indicates the company is able to make more sales of their product. Cash and cash equivalents is another powerful indicator for fundamental analysis since a company with large and increasing liquidity is better positioned to handle any headwinds, avoid bankruptcy, and seize investment opportunities to leverage growth.

#Data Cleaning#
We want five datasets that represent the stock data of small cap stocks with evaluations between \$300M and \$2B dollars market cap. For each year, we filter for the column variables we are interested in. There are quite a few variables now, but this will be narrowed down to a more narrow subset later.

```python
# initial point is 2014, so we first filter this DF for desired rows & cols

# select only small cap stocks by taking companies with valution 300M<=x<=2B
print("2014 data before filtering for market cap:", data_2014.shape)
data_2014 = data_2014[(data_2014['Market Cap'] >= 300000000.0) & (data_2014['Market Cap'] <= 2000000000.0)]
#data_2014 = data_2014[(data_2014['Market Cap'] >= 2000000000.0)]
print("2014 data after filtering for market cap:", data_2014.shape)

# filter for the columns we're interested in
data_2014 = data_2014[['Ticker','Revenue', 'R&D Expenses', 'Free Cash Flow', 
                       'Operating Cash Flow', 'Total liabilities', 'Debt to Equity',
                       'Gross Margin', 'Cash and cash equivalents', 'Sector', 
                       'Market Cap', 'Net Income', 'Earnings before Tax']]
data_2015 = data_2015[['Ticker','Revenue', 'R&D Expenses', 'Free Cash Flow', 
                       'Operating Cash Flow', 'Total liabilities', 'Debt to Equity',
                       'Gross Margin', 'Cash and cash equivalents', 'Sector', 
                       'Market Cap', 'Net Income', 'Earnings before Tax']]
data_2016 = data_2016[['Ticker','Revenue', 'R&D Expenses', 'Free Cash Flow', 
                       'Operating Cash Flow', 'Total liabilities', 'Debt to Equity',
                       'Gross Margin', 'Cash and cash equivalents', 'Sector', 
                       'Market Cap', 'Net Income', 'Earnings before Tax']]
data_2017 = data_2017[['Ticker','Revenue', 'R&D Expenses', 'Free Cash Flow', 
                       'Operating Cash Flow', 'Total liabilities', 'Debt to Equity',
                       'Gross Margin', 'Cash and cash equivalents', 'Sector', 
                       'Market Cap', 'Net Income', 'Earnings before Tax']]
data_2018 = data_2018[['Ticker','Revenue', 'R&D Expenses', 'Free Cash Flow', 
                       'Operating Cash Flow', 'Total liabilities', 'Debt to Equity',
                       'Gross Margin', 'Cash and cash equivalents', 'Sector', 
                       'Market Cap', 'Net Income', 'Earnings before Tax']]

print("\nData before dropping null values")
print("2014:", data_2014.shape)
print("2015:", data_2015.shape)
print("2016:", data_2016.shape)
print("2017:", data_2017.shape)
print("2018:", data_2018.shape)
```

Afterwards, we cleaned the data by removing entries that contained null values, and recreated our datasets by including stocks that were present in all five datsets. We then checked all the entries besides the Ticker variable were in float datatypes. At the end, we confirmed that we had five dataset containing data for the same stocks with entries that were all float values for small market cap stocks between years 2014 to 2018.

```python
# remove any entries with a null value
data_2014 = data_2014.dropna()
data_2015 = data_2015.dropna()
data_2016 = data_2016.dropna()
data_2017 = data_2017.dropna()
data_2018 = data_2018.dropna()

#data_2017 = data_2017[data_2017['Revenue'] > 0]
#data_2018 = data_2018[data_2018['Revenue'] > 0]

# match dataframes by ticker indices present in market cap restricted (2014) dataset
data_2014 = data_2014[(data_2014['Ticker'].isin(data_2015['Ticker'])) & (data_2014['Ticker'].isin(data_2016['Ticker'])) & (data_2014['Ticker'].isin(data_2017['Ticker'])) & (data_2014['Ticker'].isin(data_2018['Ticker']))]
data_2015 = data_2015[((data_2015['Ticker'].isin(data_2014['Ticker'])) & (data_2015['Ticker'].isin(data_2016['Ticker']))) & ((data_2015['Ticker'].isin(data_2017['Ticker'])) & (data_2015['Ticker'].isin(data_2018['Ticker'])))]
data_2016 = data_2016[((data_2016['Ticker'].isin(data_2014['Ticker'])) & (data_2016['Ticker'].isin(data_2015['Ticker']))) & ((data_2016['Ticker'].isin(data_2017['Ticker'])) & (data_2016['Ticker'].isin(data_2018['Ticker'])))]
data_2017 = data_2017[((data_2017['Ticker'].isin(data_2014['Ticker'])) & (data_2017['Ticker'].isin(data_2016['Ticker']))) & ((data_2017['Ticker'].isin(data_2015['Ticker'])) & (data_2017['Ticker'].isin(data_2018['Ticker'])))]
data_2018 = data_2018[(data_2018['Ticker'].isin(data_2014['Ticker'])) & (data_2018['Ticker'].isin(data_2016['Ticker'])) & ((data_2018['Ticker'].isin(data_2017['Ticker'])) & (data_2018['Ticker'].isin(data_2015['Ticker'])))]

# check that all columns (other than 'Ticker') are numeric
print("Check all data types are Floats:")
print("2014:", all(data_2014.drop(['Ticker', 'Sector'], inplace=False, axis=1).dtypes == float))
print("2015:", all(data_2015.drop(['Ticker', 'Sector'], inplace=False, axis=1).dtypes == float))
print("2016:", all(data_2016.drop(['Ticker', 'Sector'], inplace=False, axis=1).dtypes == float))
print("2017:", all(data_2017.drop(['Ticker', 'Sector'], inplace=False, axis=1).dtypes == float))
print("2018:", all(data_2018.drop(['Ticker', 'Sector'], inplace=False, axis=1).dtypes == float))

# check shapes to make sure they're all identical
print("\nFINAL DATAFRAMES:")
print("2014 shape:", data_2014.shape)
print("2015 shape:", data_2015.shape)
print("2016 shape:", data_2016.shape)
print("2017 shape:", data_2017.shape)
print("2018 shape:", data_2018.shape)
```

Lastly, we merged these five datasets into one large dataset and assigned each entry its own Year label so we could distinguish what year the data entries were from. This merged DataFrame will be helpful in plotting the raw data and observing trends in variables over time. We will then use conclusions drawn from this to guide our research in selecting effective indicator variables. 

```python
data_2014_mod = pd.DataFrame.copy(data_2014)
data_2014_mod['Year'] = 2014
data_2015_mod = pd.DataFrame.copy(data_2015)
data_2015_mod['Year'] = 2015
data_2016_mod = pd.DataFrame.copy(data_2016)
data_2016_mod['Year'] = 2016
data_2017_mod = pd.DataFrame.copy(data_2017)
data_2017_mod['Year'] = 2017
data_2018_mod = pd.DataFrame.copy(data_2018)
data_2018_mod['Year'] = 2018
all_dfs = [data_2014_mod, data_2015_mod, data_2016_mod, data_2017_mod, data_2018_mod]
merged_df = pd.concat(all_dfs)
merged_df = merged_df.sort_values(["Ticker", "Year"], ascending=True)
print("MERGED_DF:")
merged_df
```

Next, we construct a change_df DataFrame that contains proportions of all entries in 2018 divided by the corresponding entry in 2014. The resulting change_df contains float values that tell us how the variables have changed (as a growth multiple proportion) in terms of their initial values. When performing fundamental analysis, we typically want to track growth trends over a historical period. For example, it doesn't make sense to compare raw measurements of revenue between two companies. Rather, we would want to compare the growth rate of revenue over a few years with respect to revenue. To achieve this desired behavior, the change_df provides us with these more informative growth multiples while allowing for a sort of dimensionality reduction since distinct years are no longer a variable (we just have the growth over 2014-18 rather than the raw measurements of a financial metric at each of these years).

```python
# Setting change_df index to ticker
change_df = pd.DataFrame.copy(data_2014)
change_df = change_df.sort_values(['Ticker'], ascending=True)
change_df.set_index('Ticker', inplace=True)
data_2014_tindex = pd.DataFrame.copy(data_2014.drop(['Sector'], inplace=False, axis=1)).sort_values(['Ticker'], ascending=True)
data_2014_tindex.set_index('Ticker', inplace=True)
data_2018_tindex = pd.DataFrame.copy(data_2018.drop(['Sector'], inplace=False, axis=1)).sort_values(['Ticker'], ascending=True)
data_2018_tindex.set_index('Ticker', inplace=True)

change_df = (data_2018_tindex[data_2018_tindex.columns]/data_2014_tindex[data_2014_tindex.columns])
change_df = change_df.replace([np.inf, -np.inf], np.nan)
change_df = change_df.dropna()
print("CHANGE_DF:")
change_df
```

# Conclusion & Discussion

Our group was initially interested in using a company's fundamental technical variables to try and predict the company's growth as measured by growth in market cap. However, we realized the ethical complications behind this and were reminded that establishing correlation would not be enough to claim causation. As a result, we decided to explore if there was any relationship between a company's fundamental technical variables and the company's growth in market cap. Our initial sentiment, as explained in the Hyptothesis, was that there would be positive relationships between market cap growth and the fundamental variables of revenue growth and cash and cash equivalent growth because of the importance of these two variables to a company's long term health. We collaboratively picked ten promising fundamental variables, revenue, gross margin, earnings before tax, net income, debt to equity, cash and cash equivalents, total liabilities, free cash flow, operating cash flow, and research and development expenses, and ran OLS regression tests on each of these variables as related to market cap growth.

After running OLS regression on all ten of these variables, we found that only four of them to have a significant p-value. These were Revenue, Cash and cash equivalents, Total liabilities, and R&D Expenses. We decided to use these four variables to generate a Huber Regression model. We chose a Huber Regression model because this model assigned lower weights to outliers, allowing the model to be more resistant to outliers. Because of how volatile small market cap companies tend to be, it was important to restrict anomalies from impacting our results. To gauge the accuracy of the model, we extracted key accuracy metrics such as the R^2 score, Mean Squared Error, Mean Absolute Error, and Median Absolute Error. We observed an R^2 score of .56, establishing a moderate postive correlation. Additionally, we saw a Mean Squared Error of about .64, a Mean Absolute Error of about .63, and a Median Absolute Error of about .48. 

While our R^2 score of .56 seems to establish a positive relationship between the variables Revenue, Cash and cash equivalents, Total liabilities, and R&D Expenses and the growth in market cap, our error metrics are too large to claim any relationship. Taking a look at the Median Absolute Error for example, we see that the median error between our test dataset and predicted value was about .48. This means that the prediction was wrong by a margin of 48%. When it comes to investments, a 48% return is extremely substantial. As a result, we concluded that there was no clear relationship between a company's fundamental analysis variables and the company's market cap growth, thereby rejecting our hypothesis.




**Limitations**
1.   Our data set after filtering for valid small market cap companies was 225. Since our change_df DataFrame was created by taking the proportion of metrics from 2018 over 2014, the companies that had values of 0 in 2014 had to be removed from the dataset since we can't divide by zero. Even if a company was able to report a non-zero figure in the time of 2015-18, these companies were removed from our analysis for the sake of simplicity. This meant that we were only able to analyze companies with non-zero figures even though companies with zero-valued figures are not necessarily bad investments to analyze. There are many other small market cap companies that were not included in this dataset, and having more data points would lead to more accurate conclusions. It also means that we cannot apply our analysis on companies with zero-valued metrics.
2.   This project is only relevant to publicly traded companies since those were the only companies who released their complete financial details. There are many promising early stage startups that we were unable to study since they do not typically disclose their financial statements. But of course, investing in startups requires one to be an accredited investor which is out of the scope of the intended purpose of our research.
3.   There was high amounts of association among our variables, potentially negatively impacting correlation between multiple variables. The nature of a company's fundamentals is that certain metrics such as an increase in revenue typically means an increase in net income. Since these variables can be associated with one another, it may have weakened our ability to build a multi-variate regression model.
4.   Stock fundamentals are only a small part of fundamental analysis, meaning that incredibly strong growth in financial metrics is not always followed by growth in market cap. On the other hand, a company may experience strong growth in market cap even with poor fundamentals, such as the infamous trading of \$GME, which experienced extraordinary growth without much change in fundamentals. The reason is that valuations of a company come down to the balance of buying and selling pressure, and these pressures can be affected by human factors such as hype or a compelling story behind the company's vision. This means that our research is only a small part of a larger picture, and in addition to considering the financial metrics in our research, investors must consider factors such as the board members, talent aquisition, corporate reinvestment, etc.
5.   Over a 5-year period, there are external factors that may have impacted the market cap growth of our companies regardless of the underlying financial metrics. The most significant market events are the presidential election of 2016 and the Fed rate hike of 2018. These events introduced great uncertainty in the market, as can be observed in the spikes of $25 of the \$VIX volatility index. This uncertainty creates selling pressure in the market as investors de-risk and move to safer securities such as cash, bonds, and precious metals. Such external factor may weaken the correlation between a company's underlying metrics and their public valuation.
