## Height of NBA Players

Before constructing plot for player's hegihts, we need to select our data within specific draft_year range. Since the type of the draft_year is string, we created a function to standardize the type of years to be integer and replace "undrafted" to be "nan". Finally, we sliced the data from 2002 (included) to 2019 and dropped the null data and some duplicated data.

```python
def standardize_year(str_in):
    str_in = str_in.strip()
    if str_in.isnumeric():
        output = int(str_in)
    else:
        output = np.nan
    return output
```

```python
nba_height['draft_year'] = nba_height['draft_year'].apply(standardize_year)
nba_height = nba_height[nba_height['draft_year'] > 2001]
nba_height = nba_height.dropna(subset = ['draft_year','player_height'])
nba_height = nba_height.reset_index(drop=True)
nba_height.rename(columns={'player_name':'Name'}, inplace=True)
nba_height = nba_height[['Name', 'player_height']]
nba_height = nba_height.drop_duplicates(subset=None, keep='first', 
                           inplace=False, ignore_index=False).reset_index(drop=True)
nba_height.head()
```

## Combine Positions into 3 types (G, F, C)

The nba player's stats dataset provides the average FGM and average PER of each individual player from 2002 to 2019, it also shows the position of each player. We first want to remove any missing entries. After that, we want to clean the posiiton names into 3 catagories, which are F as forward, C as center, and G as guard. Thus we can use these 3 catagories to minimize the bias

```python
def clean_position(string):
    
    string = string.strip()
    
    if 'SG' in string:
        output = 'G'
    elif 'PG' in string:
        output = 'G'
    elif 'G' in string:
        output = 'G'
    elif 'SF' in string:
        output = 'F'
    elif 'PF' in string:
        output = 'F'
    elif 'F' in string:
        output = 'F'
    elif 'GF' in string:
        output = 'F'
    elif 'C' in string:
        output = 'C'
    else:
        output = string
        
    return output
```

```python
nba_stats = nba_stats.dropna()
nba_stats['Position'] = nba_stats['Position'].apply(clean_position)
```

### OLS Regression

In order to further determine whether there is a statistically significant relationship between height and achievement metrics within each position, we'll perform linear regression on players' height and their achievement metrics (FGM and PER) within each position (C, G, and F).

#### FGM ~ Position C height

```python
# carry out linear regression between height and FGM
df = df_C[['height', 'FGM']]
df.columns = ['height', 'FGM']

outcome, pred = patsy.dmatrices('FGM~height', df)
mod = sm.OLS(outcome, pred)
res = mod.fit()
print(res.summary())
```

####  FGM ~ Position G height

```python
df = df_G[['height', 'FGM']]
df.columns = ['height', 'FGM']

outcome, pred = patsy.dmatrices('FGM~height', df)
mod = sm.OLS(outcome, pred)
res = mod.fit()
print(res.summary())
```

####  FGM ~ Position F height

```python
df = df_F[['height', 'FGM']]
df.columns = ['height', 'FGM']

outcome, pred = patsy.dmatrices('FGM~height', df)
mod = sm.OLS(outcome, pred)
res = mod.fit()
print(res.summary())
```

####  PER ~ Position C height

```python
df = df_C[['height', 'PER']]
df.columns = ['height', 'PER']

outcome, pred = patsy.dmatrices('PER~height', df)
mod = sm.OLS(outcome, pred)
res = mod.fit()
print(res.summary())
```

####  PER ~ Position G height

```python
df = df_G[['height', 'PER']]
df.columns = ['height', 'PER']

outcome, pred = patsy.dmatrices('PER~height', df)
mod = sm.OLS(outcome, pred)
res = mod.fit()
print(res.summary())
```

####  PER ~ Position F height

```python
df = df_F[['height', 'PER']]
df.columns = ['height', 'PER']

outcome, pred = patsy.dmatrices('PER~height', df)
mod = sm.OLS(outcome, pred)
res = mod.fit()
print(res.summary())
```

Based on the p-values returned by the linear regression for each position, with dependent variable being either PER or FGM, we find that the only time height becomes a statistically significant predictor of career achievements is "PER ~ Position F height".  

The positive linear relationship here is that for every one unit increase in a F-position player's height, PER inceases by 0.135. The p-value for this case is very small (0.012), which suggest that we would see these results, or something more extreme due to random chance alone approximately 1.2% of the time. However, the effect size here is relatively small. This could mean that this analysis is likely failing after considering the effect of confounders, limiting our confidence in this regression result. Besides this, all the other p-values are not considered statistically significant (p-value > 0.05), meaning there is no linear relationship in all the other cases.

## Effect of Confounding Variable: Player Wingspan

The NBA Player Stat 2021-2022 and NBA Player Wingspan datasets were merged so that we could have a better understanding on the effect of potential confounding variable wingspan on player achievement matrics (PER and FGM). After merging, entries with NAN were observed and were removed from the dataset. Only one season of data was used to study the effect of wingspan on player achievement matrices because only data of current NBA players' wingspan were found at the moment. We are looking to improve this section by searching for a more complete dataset.

To study the effect of wingspan, we decided to do a linear regression between player height and PER, in addition to a linear regression between height and FGM, based on the data from season 2021-2022. Then, we would do a multivariable regression with player height and wingspan as predictors, and PER or FGM as outcomes. By comparing slopes, p-values, R-squared, and adj.R-squared, we should be able to see if wingspan is an outlier and to what extent does wingspan effect the predicting model.

### Further Exploration on Confounding Variable: Player Wingspan

In the previous section, two linear regression models were conducted. One with player height as the predictor while the other one with player height and wingspan as predictors. By comparing the two linaer regression models, we came up with the conclusion that player height is not a strong predictor for player achievement (player log(PER)). More over, the two-variable linear regression showed that our hypothesized potential confounding variable, player wingspan, could be a better predictor for player log(PER) than player height. It would be interesting to see if this conclustion would still hold true when we changed the outcome variable from player log(PER) to player(PER). Before a multivariable regression with player height and wingspan as predictors and player PER as the outcome was conducted, however, player height and wingspan was first standardized. Standardization was applied so that we could be more confident with the result when making conclusion by comparing the coefficient of each predictor.

It should be noticed that the data for player PER was not normally distributed but skewed to the right , as shown in the previous section. Player PER distribution was also shown below for quick reference.

```python
sns.histplot(wingspan_stat['PER'], bins=18, kde=True)
plt.xlabel('player PER')
plt.title('Histogram of Player PER 2021-2022')
```

```python
wingspan_stat.head()
```

```python
def standardize_value(x, average, std):
    # calculate the standardized value using (value - average) / standard deviation
    return (x - average) / std

height_mean = sum(list(wingspan_stat['height'])) / wingspan_stat.shape[0]
height_std = np.std(list(wingspan_stat['height']))
wingspan_mean = sum(list(wingspan_stat['wingspan'])) / wingspan_stat.shape[0]
wingspan_std = np.std(list(wingspan_stat['wingspan']))

wingspan_stat['std_height'] = wingspan_stat['height'].apply(standardize_value, args=(height_mean, height_std))
wingspan_stat['std_wingspan'] = wingspan_stat['wingspan'].apply(standardize_value, args=(wingspan_mean, wingspan_std))
wingspan_stat.head()
```

In order to standardize player height and wingspan, a helper function 'standardize_value' was first defined. Player height mean, player height standard deviation, player wingspan mean, player wingspan standard deviation were then calculated. Now that all the required values were obtained, standardized height and wingspan were calcualted and added to 'wingspan_stat' as new columns 'std_height' and	'std_wingspan'.

```python
outcome,predictors = patsy.dmatrices('PER ~ std_height + std_wingspan', wingspan_stat)
mod_log = sm.OLS(outcome,predictors)
res_log = mod_log.fit()
print(res_log.summary())
```

This regression result does not dispute our finding in the previous section. It showed that player height is not only a weak predictor for player log(PER), but also a weak predictor for player PER. Standardized player height as a predictor for player PER has a p-value of 0.363, meaning there is a 36.3% chance that the difference in player PER we observed is due to random chance. In addition, 95% confident interval of standardized player height includes zero, meaning the coefficent can be zero, negative, or very small. This regression result also shows that player wingspan is a stronger predictor of both player log(PER) and PER. A p-value of 0.044 indicates that the difference in player PER only has a 4.4% chance of due to random chance. 95% confident interval of standardized player wingspan does not include zero, in contrast to that of standardized player height. More importantly, since player height and wingspan used in this regression are standardized values, coefficient are a relatively strong indicator of how impactful a predictor is toward the outcome. With a coefficient of 1.6040 comparing to 0.7208, player wingspan is showed to be a stronger predictor than player height for player achievement (player PER).

# Conclusion & Discussion

Upon our analysis using the datasets we collected, we were able to draw the conclusion that there is a statistically significant relationship between the height of players in Forward position and only player efficient rating (PER) of the player career achievement metrics we defined, which were valued by player FGM (average field goal made) and player PER (player efficient rating). Meanwhile, the height of players in the other two positions (guard and center) seemed to have no effect on their achievement metrics. In this study, single/multivaribale OLS regression was used for our analysis.

Despite getting a statistically significant p-value of 0.011 at the forward positon when the outcome is player PER, there were a number of concerning statistics in the OLS regression model. First, the effect size was small. This could mean that this analysis would likely fail after considering the effect of confounders, which lowered our confidence in this regression result. In addition, the R-squared value was small. A R-squared value of 0.018 indicated that only 1.8% of the variance in the outcome player PER can be explained by player height, which further lowered our confidence in this regression result.

Our result from supplementary analysis on potential confounding variable, player wingspan, showed that our finding between NBA forwards' height and their efficeint rating was likely due to confounding variables. Using a multivariable OLS regression model with both player height and player wingspan as the predictors and player log(PER) as the outcome, we saw that correlation between player height and player log(PER) was no longer statistically significant with a p-value 0.363, while p-value of player wingspan is 0.044. This discovery made us believe that player wingspan may actually be a confounding variable and a better predictor for player PER.

In order to obtain further proves that show player wingspan as a better predictor for player PER, our data on player height and wingspan was standardized. We then used the standardized data as predictors instead of actual data, player PER as outcome instead of player log(PER), to conduct a multivariable OLS regression model. Standardizing player height and wingspan should increase our confidence in this regression result. The regression result gave a high effect size between player standardized wingspan and player PER, with a statistically significant p-value of 0.044, indicating that player wingspan is a better predictor of player PER.
