# Hypothesis


We hypothesize that Player Efficiency Rating, or PER, is the best indicator for earning a contract extension due to being a composite stat of a players impact on the court through counting stats. Measuring impact is often tricky due to variation in role, team situation and coaching, but using PER allows us to measure other aspects of the game like defensive impact and efficiency, rather than only factoring in offense. Furthermore, PER adjusts for league measures in pace and offensive/defensive ratings, which allows you to compare a whole number across years one to one, as they are standardized. This way, a PER of 15, which is considered league average, would be considered league average in 1970 and also 2022, despite drastically different pace, rules and league skillsets.

What happened here was that these rookies actually did make it past their 4th year, but they didn't resign with their original teams, at least at the time that this dataset was made, however all of these players still play in the NBA and got their contracts extended eventually or got traded elsewhere and then got another contract.

Let's edit their resign status to properly reflect that and move them to the right dataset.

```python
# Let's make a quick function that lets us edit the signed status of a player on the go
def signed(int, result):
    if result == 'yes':
        if int == 0:
            return 1
        elif int == 1:
            return int
        else:
            return int
    if result == 'no':
        if int == 1:
            return 0
        elif int == 0:
            return int
        else:
            return int
```

```python
# We want to filter out any outliers from the unsigned rookies dataset to get a better comparison
outliers_df = unsigned_rookies[unsigned_rookies['Yrs'] > 4]
outliers_df
```

```python
outliers_df['ReSigned?'] = outliers_df['ReSigned?'].apply(signed,result = 'yes')
resigned_rookies = pd.concat([resigned_rookies, outliers_df])
```

```python
# We updated resigned rookies we also need to update the unsigned rookies as we want to take out those that we switched to signed
unsigned_rookies = unsigned_rookies[unsigned_rookies['Yrs'] < 4]
unsigned_rookies.sort_values('Yrs', ascending=False)
```

A good thought excercise would be to consider two players that were drafted with similar picks, one signed and one not resigned, like Bam Adebayo (Pick 14) and Justin Jackson (Pick 15). These players were selected back to back, but experienced different results in the pros. While much of it can be attributed to how much more productive Bam was in the league, you could also speculate that system and situation had a lot to play into this.

What we can do is extrapolate each player to per48 minute metrics, which simulates if everyone had the same level of faith given to them to play 48 minutes, which is the number of minutes in a full game and compare there.

However, this is often not a great approach for evaluating players in limited minute settings, as extrapolating statistics for players with low minutes is often not how it scales, since added minutes brings more fatigue, and also yields different settings. For example, it would be considerably harder to score in the final minute of a tied game in overtime, than scoring up 40 points in garbage time where defense is less stringent.

Thats why combining for pace and minutes tends to do the best job of evaluating player performance, but it still is not failsafe. Minute projections also don't account for conditioning, and likely do not factor too much into whether a player gets re-signed or not, since teams understand the non-linear projection aspect of actually playing.

What we do see from looking at who gets re-signed or not is often a function of sunk cost. When a team spends a lot of draft capital on a player, for example a high pick or traded asset, then they are more likely to invest the time needed to develop a player. Furthermore, coaching changes often lead to greater team turnover, and can affect whether a player gets re-signed. New coaching regimes want to rebuild a team with their philosophy, and would be more likely to let go of a player who was not performing.

The final aspect to consider here is the next years draft, which plays a pivotal role in whether a player gets re-signed. If a player gets drafted in the same position then their job is already in significant danger, as the new player has more guaranteed years on their contract, and have been chosen specifically with the current team context in mind. This often is a clear indication of a team's faith in a player.

Overall, predicting whether a team will re-sign a player or not is a tricky job, and not an exact science. Player situation directly impacts the statistics that they put up, which then determines whether they get another contract. A huge part of the re-signing process that does not get represented in the statistics is potential and traits, which are largely intangible. Teams look at traits, and project how good a player can be, but it is largely conjecture. Due to this, it is hard to say exactly what best predicts a contract extension, but counting stats such as high PPG, and high advanced stats like PER or WS, also almost always guarantee a contract extension past a threshold. In the end, our data evaluation found that PPG was simply the best indicator, as valuable scorers are hard to find in the league. However, there will still be exceptions, such as Dwayne Bacon, who was a good scorer, but still did not get re-signed.
