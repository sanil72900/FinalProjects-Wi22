# Research Question

Is there a strong positive correlation between the number of registered and total voters in California by county in US elections and the median income of the coinciding county?

## Background & Prior Work

Everyone in our group had the privilege of learning about the importance of voting in high school and how to be informed voters. However, this is not the case for many adults as voter turnout in the US is low. Voter turnout for presidential elections hovers at about 60 percent and is even lower for non-presidential elections. There are many factors that affect voter turnout. Two major factors are the lack of political engagement in most schools and that it can be hard to vote. [1] The civics education that students receive vary and is dependent on the curriculum of the specific school they attend. Failing to instill the importance of voting and how to be an informed voter leads to voter apathy. Additionally, different states have differing policies when it comes to voting. This can make it harder for an individual to find the time to vote if there aren’t that many voting opportunities for them. Because of this, voting can be a costly activity as it requires time, skill, access to transportation and more. [3]

Voter turnout is also driven by various other factors, including income which can greatly affect the likelihood that an individual will vote. It turns out that richer people are more likely to vote than poorer people. This inequality is not fully understood, but some of it can be attributed to the unequal access in education, as higher income levels are more likely to be educated. There is also the fact that education systems in wealthier areas directly correlate to the quality of education in those areas. Schools that receive less funding have unequal access to better curriculum and teachers. This can negatively impact the way people view voting, especially because many political policies make it difficult to vote in poorer communities, with longer travel times to voting centers, restricting travel, and not offering paid time off to vote. Also, voting can be a costly activity in itself. In order to vote, one must have time, skills, information, a certain level of health, and access to transportation. There is a possibility that higher income provides people with resources that make voting easier. However, there is also the possibility that these interpretations regarding how income might affect voting point to a different conclusion, such as if voting requires taking time off of work it is hypothetically then more expensive for people who earn more (due to the fact that the opportunity cost of their time is higher). This approach is less popular to consider because richer people do vote more, usually regardless of the fact that their opportunity cost is higher. It can even be proven when looking at how families who acquire wealth then have a significantly higher chance of being involved with politics. The correlation between income and voter turnout exists, but its statistical significance has been hard to prove as isolating income as a factor has been difficult in the past. [2]

We will remove special characters from columns, and we will convert columns with numeric values to numbers (int and float). Removing speical characters will help make our dataframes consistenet and it will also make it easier to collect data from those columns. Converting the numberic values to numbers will ensure that we can do calucations with those values. 

```python
for df in electionDataframes:
    # remove * from county column in election dataframes to be consistent in names
    df['county'] = df['county'].str.replace('*','', regex=True)
    # remove , from registered_to_vote and total_votes columns in election dataframes
    df['registered_to_vote'] = df['registered_to_vote'].str.replace(',','', regex=True)
    df['total_votes'] = df['total_votes'].str.replace(',','', regex=True)
    # remove % from turnout_registered and turnout_eligible columns in election dataframes
    df['turnout_registered'] = df['turnout_registered'].str.replace('%','', regex=True)
    df['turnout_eligible'] = df['turnout_eligible'].str.replace('%','', regex=True)
    # convert numeric columns to numbers (int and float) in election dataframes
    df['registered_to_vote'] = pd.to_numeric(df['registered_to_vote'])
    df['total_votes'] = pd.to_numeric(df['total_votes'])
    df['turnout_registered'] = pd.to_numeric(df['turnout_registered'])
    df['turnout_eligible'] = pd.to_numeric(df['turnout_eligible'])

# one example of the changed election dataframes
print("total_votes type:", type(primary2020['total_votes'][0]))
print("turnout_registered type:", type(primary2020['turnout_registered'][0]))
primary2020.head()
```

We will remove speical characters from the median_income column in the income dataframw. This will ensure consistensith throughout the various daraframes. We will also convert the median_income column to integers, which will allow us to perform calcualtions with these values.  

```python
# remove , from median_income column in income dataframe
income['median_income'] = income['median_income'].str.replace(',','', regex=True)
# convert median_income column to integers in income dataframe
income['median_income'] = pd.to_numeric(income['median_income'])

print("median_income type:", type(income['median_income'][0]))
income.head()
```

Next we will be adding the type of election and year to each dataset. In order to do this, we will first establish three data frames: one with the general election dataframes, one with the primary election dataframes, and one with election years. Then we will add the type of election and year of election that each dataset is in separate columns. At the end, we gave an example dataset to show how the election_type and year columns fit into our final data. 

```python
# list of general election dataframes
generalElectionDataframes = [general2020, 
                             general2018, 
                             general2016,
                             general2014,
                             general2012,
                             general2010,
                             general2008,
                             general2006,
                             general2004,
                             general2002,
                             general2000]

# list of primary election dataframes
primaryElectionDataframes = [primary2020, 
                             primary2018, 
                             primary2016,
                             primary2014,
                             primary2012,
                             primary2010,
                             primary2008,
                             primary2006,
                             primary2004,
                             primary2002,
                             primary2000]

# list of all years for elections
electionYears = ['2020',
                 '2018',
                 '2016',
                 '2014',
                 '2012',
                 '2010',
                 '2008',
                 '2006',
                 '2004',
                 '2002',
                 '2000']

# add type of election and year column to each election dataframe
yearIndex = 0
for df in primaryElectionDataframes:
    df['election_type'] = 'Primary'
    df['year'] = electionYears[yearIndex]
    yearIndex += 1
    
yearIndex = 0
for df in generalElectionDataframes:
    df['election_type'] = 'General'
    df['year'] = electionYears[yearIndex]
    yearIndex += 1

# one example of election dataframes with added column
primary2020.head()
```

Then we combined the election date frame with the data frames which included income, organized by county. We added the median income of the coinciding county to each election data frame. This will allow us to accurately compare the election data with income by county, which will provide as a basis for our analysis. 

```python
# combine election dataframes with income dataframe by county
for df in electionDataframes:
    medianIncomes = []
    # add 'median income of coinciding county' column to each election dataframe
    for i in range(len(df)):
        county_name = df['county'][i]
        county_income = income[income.county == county_name]['median_income'].values[0]
        medianIncomes.append(county_income)
    df['county_median_income'] = medianIncomes

# one example of election dataframes with added column
general2000.head()
```

Our analysis of this data looked at the distribution of quantifiable variables (registered to vote, total votes, turnout registered, turnout eligible). However, in this graph we saw that there was a correlation in certain income ranges. Specifically, in the higher income bracket there was a negative correlation and for the middle income bracket there was a positive correlation. Next, we considered each county's median income by voter turnouts with a line plot; these graphs were broken down by the year of election. Again, there was no comprehensive correlation found. But, we did find that the general trend for each year is the same throughout the 20 years we were analyzing. We then compared county’s whose median income were less than 85,000 to voter turnouts and no correlation was found. We next looked at county’s which had median incomes, from 85,000 to 110,217, in comparison to voter turnout. Here, there was a positive correlation found. On the other hand, for counties with median incomes of above 110,217, in comparison to voter turnouts, there was a negative correlation found.

Overall, we were unable to find a comprehensively inclusive trend that showed a strong positive correlation between the number of registered and total voters in California by county in US elections in comparison to the median income of the coinciding counties. However, we did find a correlation when examining the counties which had mid to higher range incomes. We saw that mid range income counties had a positive correlation between income and turnout versus the higher income counties had a negative correlation.
